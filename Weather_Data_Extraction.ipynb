```python
import csv
import requests

CSV_URL = "https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/"+LOCATION+"/"+Start_Date+"/"+End_Date+"?unitGroup=metric&include=days&key="+API_KEY+"&contentType=csv"


with requests.Session() as s:
    download = s.get(CSV_URL)

    decoded_content = download.content.decode('utf-8')

    cr = csv.reader(decoded_content.splitlines(), delimiter=',')
    my_list = list(cr)
    for row in my_list:
        print(row)
```
